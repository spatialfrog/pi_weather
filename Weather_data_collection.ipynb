```python
while True:
    pressure = sense.get_pressure()
    pressure = round(pressure, 3)
    
    humidity = sense.get_humidity()
    humidity = round(humidity, 3)
    
    print 'Pressure: {0}\t Humidty: {1}'.format(pressure, humidity)
    
```
